# Indox Retrieval Augmentation for CSV files

in this we are using the indox retrieval augmentation to retrieve the data.

## Docx Loader

Best for: Word document text extraction with paragraph analysis

## Features

- Paragraph-based page estimation
- Style-aware text extraction
- Metadata preservation


```python
from indoxArcg.data_loaders import Docx

# Example usage:
loader = Docx(paragraphs_per_page=1)  # Customize page estimation
docs = loader.load("./data/Bitcoin.docx")
print(docs.page_content)
```

# OpenPyXl Loader

Best for: Excel spreadsheet data processing

## Features
- Multi-sheet handling
- Pandas DataFrame integration
- Cell formatting awareness


```python
# Example usage:


loader = OpenPyXl(sheet_names=["Sheet1"])
docs = loader.load("./data/fake_cancer_data.xlsx")
docs
```

# Pptx Loader

Best for: PowerPoint presentation content extraction

## Features

- Slide-level text extraction
- Speaker notes inclusion
- Slide metadata tracking


```python
from indoxArcg.data_loaders import Pptx

pptx_loader = Pptx(include_notes=True)
documents = pptx_loader.load("./data/Presentation.pptx")
for doc in documents:
    print(doc.page_content)
```
