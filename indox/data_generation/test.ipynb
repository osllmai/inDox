```python
import os
from dotenv import load_dotenv
from synthetic.synthetic_data import SyntheticData
from llms.openai import OpenAI
from prompt import DataFromPrompt
from prompt.data_generation_prompt import DataGenerationPrompt 
from upload_data import load_data

load_dotenv()
INDOX_API_KEY = os.getenv("INDOX_API_KEY")

dataset_file_path = "generated_astronomy_dataset.xlsx"

df = load_data(dataset_file_path)

user_prompt = " generate one row."
with SyntheticData("./output"):
    # Load GPT-4
    gpt_4 = OpenAI(api_key=INDOX_API_KEY, model_name="gpt-4o-mini")

    instruction = DataGenerationPrompt.get_instruction(user_prompt)

    dataset = DataFromPrompt(
        prompt_name="Generate New Column",
        args={
            "llm": gpt_4,
            "n": 1,
            "temperature": 0.1,
            "instruction": instruction,
        },
        outputs={"generations": "generate"},
        dataframe=df 
    )
    updated_df = dataset.run()
    print(updated_df)
```

```python
import os
from dotenv import load_dotenv
from llms.openai import OpenAI
from prompt import FewShotPrompt  

load_dotenv()
INDOX_API_KEY = os.getenv("INDOX_API_KEY")

examples = [
    {
        "input": "Generate a dataset with 3 columns and 2 rows about biology.",
        "output": '[{"Species": "Human", "Cell Count": 37.2, "Age": 30}, {"Species": "Mouse", "Cell Count": 3.2, "Age": 2}]'
    },
    {
        "input": "Generate a dataset with 3 columns and 2 rows about chemistry.",
        "output": '[{"Element": "Hydrogen", "Atomic Number": 1, "Weight": 1.008}, {"Element": "Oxygen", "Atomic Number": 8, "Weight": 15.999}]'
    }
]

user_prompt = "Generate a dataset with 3 columns and 2 rows about astronomy."

gpt_4 = OpenAI(api_key=INDOX_API_KEY, model_name="gpt-4o-mini")

dataset = FewShotPrompt(
    prompt_name="Generate Astronomy Dataset",
    args={
        "llm": gpt_4,
        "n": 1,  
        "instruction": user_prompt,  
    },
    outputs={"generations": "generate"},
    examples=examples  
)

df = dataset.run()

print(df)
df.to_excel("generated_astronomy_dataset.xlsx", index=False)
```
