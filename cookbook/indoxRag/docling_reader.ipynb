```python
import os
from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.environ['OPENAI_API_KEY']
```

```python
from indoxRag import IndoxRetrievalAugmentation

indox = IndoxRetrievalAugmentation()
```

```python
from indoxRag.embeddings import OpenAiEmbedding
from indoxRag.llms import OpenAi

openai_qa = OpenAi(api_key=OPENAI_API_KEY, model="gpt-4o-mini")
embed_openai = OpenAiEmbedding(api_key=OPENAI_API_KEY, model="text-embedding-3-small")

from indoxRag.vector_stores import Chroma
db = Chroma(collection_name="sample",embedding_function=embed_openai)
```

```python
file_path = "samp.pdf"

from indoxRag.data_loaders import DoclingReader
```

```python
docling = DoclingReader(file_path=file_path)
docs = docling.load()
```

```python
from docling.document_converter import DocumentConverter
from docling.chunking import HybridChunker
```

```python
converter = DocumentConverter()
docs = converter.convert(source="samp.pdf")
```

```python
text_docs = docs.document.export_to_text()
```

```python
print(docs.document.export_to_markdown())
```

```python
from indoxRag.splitter import SemanticTextSplitter
splitter = SemanticTextSplitter(chunk_size=400)
chunks = splitter.split_text(text_docs)
```

```python
chunks[0]
```

```python
db.add(docs=chunks)
```

```python
query = "introduce docling to in a simplest way"
retriever = indox.QuestionAnswer(vector_database=db, llm=openai_qa, top_k=5)
```

```python
answer = retriever.invoke(query)
context = retriever.context
```
